```python
import scipy.stats as stats
def plotting(data,var):
    plt.subplot(1,3,1)
    plt.hist(data[var])
    plt.subplot(1,3,2)
    stats.probplot(data[var],dist="norm",plot=plt)
    plt.subplot(1,3,3)
    sns.boxplot(data[var])
plotting(data,"Age")
```

```python
def correlations(dataset,threshold):
    col=set()
    matrix=dataset.corr()
    for i in range(len(matrix.columns)):
        for j in range(i):
            if(matrix.iloc[i,j]>threshold):
                col_na=matrix.columns[i]
                col.add(col_na)
    return col
feat=correlations(x,0.4)  
feat
```
