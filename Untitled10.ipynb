```python
miss_data=[i for i in data.columns if data[i].isnull().sum()>0]
```

```python
num_data=[i for i in data.columns if data[i].dtype!="O"]
```

```python
data[miss_data].isnull().mean()
```

```python
cat_data=[i for i in data.columns if data[i].dtype=="O"]
```

```python
cat_data
```

```python
plt.figure(figsize=(25,20))
sns.heatmap(data[num_data].corr(),annot=True)
```

```python
disc=[i for i in num_data if len(data[i].unique())<20]
```

```python
disc
```

```python
cont=[i for i in num_data if i not in disc]
```

```python
cont
```

```python
from scipy import stats
for i in cont:
    plt.subplot(1,3,1)
    plt.title(i)
    sns.boxplot(data[i])

    plt.subplot(1,3,2)
    plt.title(i)
    sns.histplot(data[i])
    plt.show()
    plt.subplot(1,3,3)
    plt.title(i)
    stats.probplot(data[i],dist="norm",plot=plt)
    plt.show()
```

```python
for i in num_data:
    if data[i].dtype!="0":
        data[i]=data[i].fillna(data[i].median())
    
```

```python
data[num_data].isnull().sum()
```

```python
for i in cat_data:
    if data[i].dtype=="O":
        data[i]=data[i].fillna(data[i].mode()[0])
```

```python
data[cat_data].isnull().sum()
```
