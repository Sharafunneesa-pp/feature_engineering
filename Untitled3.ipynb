```python
def correlations(dataset,threshold):
    col=set()
    matrix=dataset.corr()
    for i in range(len(matrix.columns)):
        for j in range(i):
            if(matrix.iloc[i,j]>threshold):
                col_na=matrix.columns[i]
                col.add(col_na)
    return col
feat=correlations(num_train,0.8)  
feat
```
