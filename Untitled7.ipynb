```python
sns.heatmap(data[numeric_num].corr(),annot=True)
```

```python
from scipy import stats
for i in con_num:
    plt.subplot(1,3,1)
    plt.title(i)
    sns.histplot(data[i])
    plt.subplot(1,3,2)
    stats.probplot(data[i],dist="norm",plot=plt)
    plt.subplot(1,3,3)
    sns.boxplot(data[i])
    plt.show()
    
```

```python
for var in desc_num:
    (data.groupby(var)[var].count()/len(data)).plot.bar()
    plt.show()
```

```python
for var in desc_num:
    data.groupby(var)["SalePrice"].median().plot()
```

```python
data=ct.fit_transform(data)
```
